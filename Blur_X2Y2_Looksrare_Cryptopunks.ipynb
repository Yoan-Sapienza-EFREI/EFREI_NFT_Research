```python
def Drop_dup(path_file, name):
    unique_dicts = set([])
    with open(path_file, 'r') as input_file, open(f"/Users/zappy/Documents/Marketplace_NFT_Transactions/{name}.json", 'w') as output_file:
        for line in input_file:
            # Convertir la chaîne JSON en dictionnaire
            current_dict = json.loads(line)

            # Ajouter le dictionnaire à l'ensemble de dictionnaires uniques
            unique_dicts.add(json.dumps(current_dict))
            if len(unique_dicts) == 900:
                for unique_dict in unique_dicts:
                    output_file.write(unique_dict + '\n')
                unique_dicts.clear()
                
Drop_dup(path, "data_blur_Clean")
```
